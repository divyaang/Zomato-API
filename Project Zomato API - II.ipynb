## 1.1: Plot the bar graph of number of restaurants present in Delhi NCR vs Rest of India

## 1.3: Find the top 10 cuisines served by maximum number of restaurants in Delhi NCR and rest of India

```python
df = pd.read_csv('zomato.csv', encoding='latin-1')
df_ind = df[df['Country Code'] == 1]
```

```python
cuisine_list = list(df_ind['Cuisines'])
city_list = list(df_ind['City'])
ncr_list = ['New Delhi', 'Ghaziabad', 'Noida', 'Gurgaon', 'Faridabad']

ncr = dict()
roi = dict()

for i in range(len(city_list)):
    if city_list[i] in ncr_list:
        cl=cuisine_list[i].split(',')
        for c in cl:
            c = c.strip()
            ncr[c] = ncr.get(c,0) + 1
    else:
        cl=cuisine_list[i].split(',')
        for c in cl:
            c = c.strip()
            roi[c] = roi.get(c,0) + 1
            
ncr_key = np.array(list(ncr.keys()))
ncr_value = np.array(list(ncr.values()))
roi_key = np.array(list(roi.keys()))
roi_value = np.array(list(roi.values()))

ncr_ind = ncr_value.argsort()[::-1]
ncr_ind = ncr_ind[:10]

roi_ind = roi_value.argsort()[::-1]
roi_ind = roi_ind[:10]

ncr_key = ncr_key[ncr_ind]
ncr_value = ncr_value[ncr_ind]

roi_key = roi_key[roi_ind]
roi_value = roi_value[roi_ind]

print('Top 10 cuisines served by maximum number of restaurants in Delhi NCR are:')
for i in range(10):
    print(i+1,ncr_key[i],'-',ncr_value[i])

print()
print('Top 10 cuisines served by maximum number of restaurants in Rest of India are:')
for i in range(10):
    print(i+1,roi_key[i],'-',roi_value[i])
    
plt.bar(ncr_key,ncr_value)
plt.ylabel('Number of Restaurants',fontsize=15)
plt.title('Top 10 cuisines served by maximum number of restaurants in Delhi NCR',fontsize=15)
plt.yticks(fontsize=15)
plt.xticks(rotation='vertical',fontsize=15)
plt.show()

plt.bar(roi_key,roi_value)
plt.ylabel('Number of Restaurants',fontsize=15)
plt.title('Top 10 cuisines served by maximum number of restaurants in Rest of India',fontsize=15)
plt.yticks(fontsize=15)
plt.xticks(rotation='vertical',fontsize=15)
plt.show()

```

## 2.1.2: User Rating Versus Number of Cuisines 

```python
from pylab import rcParams
rcParams['figure.figsize'] = 8,5

cuisine = list(df_ind['Cuisines'])
rating = list(df_ind['Rating color'])
cuisine_count = []

for i in range(len(cuisine)):
    cl=cuisine[i].split(',')
    cuisine_count.append(len(cl))

plt.scatter(rating,cuisine_count)
plt.ylabel('Number of Cuisines',fontsize=15)
plt.xlabel('User Rating',fontsize=15)
plt.title('User Rating Versus Number of Cuisines',fontsize=15)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.show()
    
```

```python
rating = list(df_ind['Aggregate rating'])
cui = list(df_ind['Cuisines'])

d={}

for i in range(len(cui)):
    cl = cui[i].split(',')
    l = len(cl)
    if l in d:
        li = d[l]
        li.append(rating[i])
        d[l] = li
    else:
        d[l] = [rating[i]]
        
for k, v in d.items():
    d[k] = sum(v)/len(v)
    
count = np.array(list(d.keys()))
rate = np.array(list(d.values()))
rate = np.round(rate, decimals=2)

for i in range(len(rate)):
    print('Average User Rating of Restuarants serving',count[i],'Cuisines is',rate[i])

plt.bar(count,rate)
plt.ylabel('Average User Rating',fontsize=15)
plt.xlabel('Number of Cuisines Served',fontsize=15)
plt.title('Average User Rating Versus Number of Cuisines Served',fontsize=15)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.show()
```

## 2.1.4: User Rating Versus Restaurant serving some specific cuisines

```python
rating = list(df_ind['Aggregate rating'])
cui = list(df_ind['Cuisines'])

d={}

for i in range(len(cui)):
    cl = cui[i].split(',')
    for c in cl:
        c = c.strip()
        if c in d:
            l = d[c]
            l.append(rating[i])
            d[c] = l
        else:
            d[c] = [rating[i]]
        
for k, v in d.items():
    d[k] = sum(v)/len(v)
    
cuisine = np.array(list(d.keys()))
rate = np.array(list(d.values()))
rate = np.round(rate, decimals=2)


#TOP 10 Cuisines
ind = rate.argsort()[::-1]
ind = ind[:10]

cuisine_t10 = cuisine[ind]
rate_t10 = rate[ind]

print('Top 10 Rated Cuisines are:')
for i in range(10):
    print(cuisine_t10[i],'-',rate_t10[i])
    
plt.bar(cuisine_t10, rate_t10)
plt.ylabel('Average User Rating',fontsize=15)
plt.xlabel('Cuisine',fontsize=15)
plt.title('Top 10 Rated Cuisines',fontsize=15)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15, rotation='vertical')
plt.show()

#LOWEST 10 Cuisines
ind = rate.argsort()
ind = ind[:10]

cuisine_l10 = cuisine[ind]
rate_l10 = rate[ind]

print('Lowest 10 Rated Cuisines are:')
for i in range(10):
    print(cuisine_l10[i],'-',rate_l10[i])
    
plt.bar(cuisine_l10, rate_l10)
plt.ylabel('Average User Rating',fontsize=15)
plt.xlabel('Cuisine',fontsize=15)
plt.title('Lowest 10 Rated Cuisines',fontsize=15)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15, rotation='vertical')
plt.show()
```

## 2.2.1: Find the weighted restaurant rating of each locality

```python
df = pd.read_csv('zomato.csv', encoding='latin-1')
df_ind = df[df['Country Code'] == 1]
```

```python
votes = list(df_ind['Votes'])
rating = list(df_ind['Aggregate rating'])
locality = list(df_ind['Locality'])

d={}

for i in range(len(locality)):
    if locality[i] in d:
        l = d[locality[i]]
        vr = l[0] + (votes[i]*rating[i])
        v = l[1] + votes[i]
        d[locality[i]] = [vr,v]
    else:
        vr = (votes[i]*rating[i])
        v = votes[i]
        d[locality[i]] = [vr,v]
        
for k,v in d.items():
    if v[1] != 0:
        d[k] = v[0]/v[1]
    else:
        d[k] = 0

#Weighted restaurant rating of each locality        
loca = np.array(list(d.keys()))
wr = np.array(list(d.values()))

ind = wr.argsort()[::-1]
ind = ind[:10]

#Top 10 localities with more weighted restaurant rating
loca = loca[ind]
wr = wr[ind]
wr = np.round(wr, decimals=2)

print('The Top 10 localities with more Weighted Restaurant Rating are:')
for i in range(10):
    print(i+1,loca[i],'-', wr[i])
    
```

## 3.5: Plot the bubble graph of a number of Restaurants present in the city of India and keeping the weighted restaurant rating of the city in a bubble

```python
df = pd.read_csv('zomato.csv', encoding='latin-1')
df_ind = df[df['Country Code'] == 1]
```

```python
votes = list(df_ind['Votes'])
rating = list(df_ind['Aggregate rating'])
city = list(df_ind['City'])

d={}

for i in range(len(city)):
    if city[i] in d:
        l = d[city[i]]
        vr = l[0] + (votes[i]*rating[i])
        v = l[1] + votes[i]
        c = l[2] + 1
        d[city[i]] = [vr,v,c]
    else:
        vr = (votes[i]*rating[i])
        v = votes[i]
        d[city[i]] = [vr,v,1]
        
for k,v in d.items():
    if v[1] != 0:
        d[k] = [(v[0]/v[1]),v[2]]
    else:
        d[k] = [0,v[2]]

        
cit = np.array(list(d.keys()))
x = np.array(list(d.values()))

wr=[]
c=[]

for i in x:
    wr.append(i[0])
    c.append(i[1])
    
wr = np.array(wr)
wr = np.round(wr, decimals = 3)
c = np.array(c, dtype = int)

for i in range(len(cit)):
    print(i+1,cit[i],'has',c[i],'Restaurants with weighted Restaurant Rating', wr[i])

rcParams['figure.figsize'] = 15,5
plt.scatter(cit, c, s = wr)
plt.ylabel('Number of Restaurants',fontsize=15)
plt.title('Bubble graph of a number of Restaurants present in the city of India',fontsize=15)
plt.yticks(fontsize=15)
plt.xticks(rotation='vertical',fontsize=15)
plt.show()


```
